##Renomeando colunas

```python
base_unica.info()
```

```python
base_unica_renomeada = base_unica.rename(columns={
    'Gender': 'ds_genero',
    'Age': 'nr_idade',
    'Height': 'nr_altura',
    'Weight': 'nr_peso',
    'family_history': 'is_historico_familiar',
    'FAVC': 'is_alimentacao_calorica_frequente',
    'FCVC': 'qt_vegetais_refeicoes',
    'NCP': 'qt_refeicoes_principais',
    'CAEC': 'ds_come_entre_refeicoes',
    'SMOKE': 'is_fumante',
    'CH2O': 'qt_agua_diaria',
    'SCC': 'is_monitora_caloria_diaria',
    'FAF': 'qt_dias_atividade_fisica',
    'TUE': 'qt_horas_eletronicos',
    'CALC': 'ds_frequencia_alcoolicos',
    'MTRANS': 'ds_tipo_transporte',
    'Obesity': 'ds_obesidade'
})
base_unica_renomeada.head()
```

##Traduzindo e ajustando formato dos dados

```python
base_unica_renomeada['ds_genero'].unique()
```

```python
base_unica_renomeada['ds_come_entre_refeicoes'].unique()
```

```python
base_unica_renomeada['ds_frequencia_alcoolicos'].unique()
```

```python
base_unica_renomeada['ds_tipo_transporte'].unique()
```

```python
base_unica_renomeada['ds_obesidade'].unique()
```

```python
base_traduzida = base_unica_renomeada.copy()
```

```python
base_traduzida['ds_genero'] = base_traduzida['ds_genero'].replace({'Male': 'Masculino', 'Female': 'Feminino'})
base_traduzida['ds_come_entre_refeicoes'] = base_traduzida['ds_come_entre_refeicoes'].replace({'Sometimes': 'Às vezes', 'Frequently': 'Frequentemente', 'Always': 'Sempre', 'no': 'Não'})
base_traduzida['ds_frequencia_alcoolicos'] = base_traduzida['ds_frequencia_alcoolicos'].replace({'Sometimes': 'Às vezes', 'Frequently': 'Frequentemente', 'Always': 'Sempre', 'no': 'Não'})
base_traduzida['ds_tipo_transporte'] = base_traduzida['ds_tipo_transporte'].replace({'Public_Transportation': 'Transporte Público', 'Walking': 'Caminhada', 'Automobile': 'Carro', 'Motorbike': 'Motocicleta', 'Bike': 'Bicicleta'})
base_traduzida['ds_obesidade'] = base_traduzida['ds_obesidade'].replace({'Insufficient_Weight': 'Abaixo do Peso', 'Normal_Weight': 'Peso Normal', 'Overweight_Level_I': 'Acima do Peso Nível I', 'Overweight_Level_II': 'Acima do Peso Nível II', 'Obesity_Type_I': 'Obesidade Tipo I', 'Obesity_Type_II': 'Obesidade Tipo II', 'Obesity_Type_III': 'Obesidade Tipo III'})
base_traduzida['is_historico_familiar'] = base_traduzida['is_historico_familiar'].replace({'yes': 'Sim', 'no': 'Não'})
base_traduzida['is_alimentacao_calorica_frequente'] = base_traduzida['is_alimentacao_calorica_frequente'].replace({'yes': 'Sim', 'no': 'Não'})
base_traduzida['is_fumante'] = base_traduzida['is_fumante'].replace({'yes': 'Sim', 'no': 'Não'})
base_traduzida['is_monitora_caloria_diaria'] = base_traduzida['is_monitora_caloria_diaria'].replace({'yes': 'Sim', 'no': 'Não'})

base_traduzida.head()
```

```python
base_traduzida['nr_idade'] = base_traduzida['nr_idade'].astype(int)
base_traduzida['nr_altura'] = base_traduzida['nr_altura'].round(2)
base_traduzida['nr_peso'] = base_traduzida['nr_peso'].round(1)
base_traduzida['qt_vegetais_refeicoes'] = base_traduzida['qt_vegetais_refeicoes'].astype(int)
base_traduzida['qt_refeicoes_principais'] = base_traduzida['qt_refeicoes_principais'].astype(int)
base_traduzida['qt_agua_diaria'] = base_traduzida['qt_agua_diaria'].round(1)
base_traduzida['qt_dias_atividade_fisica'] = base_traduzida['qt_dias_atividade_fisica'].astype(int)
base_traduzida['qt_horas_eletronicos'] = base_traduzida['qt_horas_eletronicos'].round(1)

base_traduzida.head()
```

```python
base_traduzida.to_csv('/content/drive/MyDrive/Cursos/Pós/Tech_challenge_4/base_tratada.csv', index=False)
```
